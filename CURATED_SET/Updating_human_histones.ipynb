```python
import pandas as pd
from curated_set_services import CuratedSet
```

```python
df = pd.read_csv("human_histones.csv",engine='python').fillna('')
df.shape, df.columns
```

```python
print("Dropping %d records"%len(df[df["RefSeq peptide ID"] == ""]))
df = df.drop(df[df["RefSeq peptide ID"] == ""].index)
```

```python
#Currently we need manually add variant group
def var2vargroup(var):
    v1=var.split(".")[0]
    if((v1=="cH4")or(v1=="cH3")or(v1=="cH2A")or(v1=="cH2B")):
        return v1
    if((var[0:5])=="H2A.Z"):
        return "H2A.Z"
    if(var=="cH4"):
        return "cH4"
```

```python
df_to_add=pd.DataFrame({"accession":df["RefSeq peptide ID"].values,
              "type":df["Histone type"].values,
              "variant_group":[var2vargroup(i) for i in df["Histone variant"].values],
              "variant":[i+"_(Homo_sapiens)" for i in df["Histone variant"].values],
              "ncbi_gene_id":df["NCBI gene ID"].values,
             "hgnc_gene_name":df["HGNC symbol"].values,
             "taxonomy_id":"9606",
             "organism":"Homo sapiens",
             "phylum":"Chordata",
             "class":"Mammalia",
             "taxonomy_group":"Mammalia",
             "info":df["Info"].values,
             "references":df["References"].values,
             "ncbi_gene_id":df["NCBI gene ID"].values},index=df["RefSeq peptide ID"].values)
df_to_add.head()
```
